```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Función para aplicar Maxgreen Gsc-CLAHE a una imagen
def apply_maxgreen_gsc_clahe(image_path):
    # Cargar la imagen en modo color
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)
    
    # Extraer el canal verde
    green_channel = img[:,:,1]
    
    # Aplicar CLAHE al canal verde
    # Crear un objeto CLAHE (el límite de contraste y tamaño de bloque se pueden ajustar)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    clahe_img = clahe.apply(green_channel)
    
    # Sustituir el canal verde en la imagen original por el canal mejorado con CLAHE
    img[:,:,1] = clahe_img

    # Convertir la imagen de BGR a RGB para mostrar con matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Mostrar la imagen original y la imagen procesada
    plt.figure(figsize=(15, 15))

    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title('Maxgreen Gsc-CLAHE Applied')
    plt.imshow(img_rgb)
    plt.axis('off')

    plt.show()

# Asumimos que la imagen se ha cargado previamente en el entorno de ejecución
image_path = 'original.jpg'  # Replace with the path of the uploaded image
apply_maxgreen_gsc_clahe(image_path)
```
